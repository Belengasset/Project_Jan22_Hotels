```python
# crear mapa

mapa=folium.Map(location=[df.latitude.mean(), df.longitude.mean()],  # la posicion
                tiles='openstreetmap',            # tipo de mapa, theme
                zoom_start=2)                    # zoom inicial

mapa
```

```python
for index, row in df_coord.iterrows():# método para iterar por los dataframes
    if row['Rank_Year'] == 2022:
        folium.Marker([row["latitude"], row['longitude']],
                      radius=10, 
                      tooltip = row['Hotel'],
                      icon=folium.Icon(icon = "home", color = "purple")).add_to(mapa)
    elif row['Rank_Year'] == 2021:
        folium.Marker([row["latitude"], row['longitude']],
                      radius=10, 
                      tooltip = row['Hotel'],
                      icon=folium.Icon(icon = "info-sign", color = "darkgreen")).add_to(mapa)
    else:
        pass
```

```python
import geopy
from geopy.geocoders import Nominatim
from time import sleep

def add_coordinates(data_frame, city_column, country_column):
    geolocator = Nominatim(user_agent="geoapiExercises")
    data_frame["latitude"] = None
    data_frame["longitude"] = None
    for i in range(len(data_frame)):
        location = geolocator.geocode(data_frame.loc[i, city_column] + ", " + data_frame.loc[i, country_column])
        if location:
            data_frame.loc[i, "latitude"] = location.latitude
            data_frame.loc[i, "longitude"] = location.longitude
        if i % 10 == 0 and i != 0:
            sleep(10)
    return data_frame
```
